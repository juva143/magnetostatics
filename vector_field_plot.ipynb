```python
import numpy as np
import matplotlib.pyplot as plt

from pyradia.parallepiped import Parallelepiped
import radia as rad
```

# B field

```python

c = [0,0,0]
w = [2,2,4]
m = [0,0,2]

x = np.linspace(-3,3,9)
z = np.linspace(-3,3,9)
X, Z = np.meshgrid(x,z)
Uown, Vown = np.zeros(X.shape), np.zeros(Z.shape)

mblock = Parallelepiped(center=c, dimensions=w, magnetization=m)

for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        p = [X[i,j],0,Z[i,j]]
        Uown[i,j], _, Vown[i,j] = mblock.calc_Bfield(r0=p)

plt.quiver(X,Z,Uown,Vown)
plt.plot([-w[0]/2,w[0]/2],2*[w[2]/2],c='blue')
plt.plot([-w[0]/2,w[0]/2],2*[-w[2]/2],c='blue')
plt.plot(2*[-w[0]/2],[-w[2]/2,w[2]/2],c='blue')
plt.plot(2*[w[0]/2],[-w[2]/2,w[2]/2],c='blue')
plt.title('parallelepiped')
plt.xlabel('x')
plt.ylabel('z')
plt.show()
```

```python
c = [0,0,0]
w = [2,2,4]
m = [0,0,2]

x = np.linspace(-3,3,9)
z = np.linspace(-3,3,9)
X, Z = np.meshgrid(x,z)
Uradia, Vown = np.zeros(X.shape), np.zeros(Z.shape)

obj = rad.ObjRecMag(c,w,m)

for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        p = [X[i,j],0,Z[i,j]]
        Uradia[i,j], _, Vown[i,j] = rad.Fld(obj,'b',p)

plt.quiver(X,Z,Uradia,Vown)
plt.plot([-w[0]/2,w[0]/2],2*[w[2]/2],c='blue')
plt.plot([-w[0]/2,w[0]/2],2*[-w[2]/2],c='blue')
plt.plot(2*[-w[0]/2],[-w[2]/2,w[2]/2],c='blue')
plt.plot(2*[w[0]/2],[-w[2]/2,w[2]/2],c='blue')
plt.title('radia')
plt.xlabel('x')
plt.ylabel('z')
plt.show()
```

## 3D quiver

```python


class VectorField3D:
    def __init__(self, x, y, z, u, v, w):
        """
        Inicializa a classe com os componentes do campo vetorial.
        
        Parmetros:
            x, y, z: Arrays numpy com as coordenadas de origem dos vetores.
            u, v, w: Arrays numpy com as componentes dos vetores em x, y e z.
        """
        self.x = x
        self.y = y
        self.z = z
        self.u = u
        self.v = v
        self.w = w

    def plot_quiver(self):
        """Gera um quiver plot 3D do campo vetorial armazenado."""
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        ax.quiver(self.x, self.y, self.z, self.u, self.v, self.w, length=0.1, normalize=True)

        # Personalizao dos eixos
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        ax.set_zlabel('Z')
        
        plt.show()
```

```python

# Exemplo de uso
# Dados para o campo vetorial
x, y, z = np.meshgrid(np.arange(-1, 1, 0.2), np.arange(-1, 1, 0.2), np.arange(-1, 1, 0.2))
u = -y
v = x
w = np.zeros_like(x)


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(x, y, z, u, v, w, length=0.1, normalize=True)

# Personalizao dos eixos
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()

```

```python
# Define the vector field function
def f(x, y, z):
    return -y, x, 0
 
# Define the grid for plotting
x = np.linspace(-1, 1, 11)
y = np.linspace(-1, 1, 11)
z = np.linspace(-1, 1, 11)
X, Y, Z = np.meshgrid(x, y, z)
 
# Calculate the vector components on the grid
VX, VY, VZ = f(X, Y, Z)
 
# Set up the plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
# Plot the 3D quiver
ax.quiver(X, Y, Z, VX, VY, VZ, length=0.2, normalize=True)
 
# Set plot limits
# ax.set_xlim([-3, 3])
# ax.set_ylim([-3, 3])
# ax.set_zlim([-3, 3])
 
plt.show()
```
