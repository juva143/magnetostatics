```python
import numpy as np
import matplotlib.pyplot as plt

import radia as rad
from pyradia.magblock import Parallelepiped, Polyhedron, Magnet
from pyradia.magplot import plot_magnetic_block
```

```python
%matplotlib qt6
```

```python
mblock1 = Parallelepiped(dimensions=[2,2,4],center=[-1,1,0.5],magnetization=[1,2,3])
```

```python
plot_magnetic_block(mblock1,edgecolor='k',xlabel='x',ylabel='y',zlabel='z',
                    xlim=(-5,5),ylim=(-5,5),zlim=(-5,5))
```

```python

vertices = [[-0.65,-0.24,3], # 0 A
            [1.45,-0.16,1],  # 1 B
            [0,2,-1],        # 2 C
            [-2,0,0],        # 3 D
            [-1,2,5]]        # 4 E


faces_indices = [
    [0,1,4],[1,2,4],
    [0,4,3],[2,3,4],
    [0,3,1],[1,3,2]
]

mblock2 = Polyhedron(vertices=vertices, faces_indices=faces_indices, magnetization=[1,2,3])
```

```python
plot_magnetic_block(mblock2, facecolors='red', linewidths=1, edgecolors='k',
                    xlim=(-5,5),ylim=(-5,5),zlim=(-2,8))
```

```python
ax = plt.subplot(projection='3d')

plot_magnetic_block(mblock1,ax=ax,show=False,edgecolor='k',xlabel='x',ylabel='y',zlabel='z',
                    xlim=(-5,5),ylim=(-5,5),zlim=(-5,5))

plot_magnetic_block(mblock2,ax=ax,show=False,facecolors='red', linewidths=1, edgecolors='k')

plt.show()
```

```python

r0 = [3,-2,4]

mblock1 = Parallelepiped(dimensions=[2,2,4],center=[4,-2,0],magnetization=[1,2,3])

ax = plt.subplot(projection='3d')
plot_magnetic_block(mblock1,ax=ax,show=False,edgecolor='k',xlabel='x',ylabel='y',zlabel='z',
                    xlim=(-5,5),ylim=(-5,5),zlim=(-5,5))
plot_magnetic_block(mblock2,ax=ax,show=False,facecolors='red', linewidths=1, edgecolor='k')
plt.plot([r0[0]],[r0[1]],[r0[2]],'.')

plt.show()
```
