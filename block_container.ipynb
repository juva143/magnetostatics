```python
import numpy as np
import matplotlib.pyplot as plt

from pyradia.parallepiped import Parallelepiped, RectangularObject
from pyradia.mag_plot import parallelepiped_plot

import radia as rad
```

```python
c = [-3,0,0]
w = [2,2,4]
m = [0,0,2]

p = [0,0,1]

mblock = Parallelepiped(c,w,m)
print(f'B field: {mblock.calc_Bfield(p).round(4)} T')

obj = rad.ObjRecMag(c,w,m)
print(f"B field: {np.round(rad.Fld(obj,'b',p),4)} T")
```

```python
parallelepiped_plot(mblock)
plt.xlim(-6,0)
plt.ylim(-3,3)
plt.show()
```

```python

w = [2,2,4]
m = [0,0,2]

p = [0,0,1]

mblock1 = Parallelepiped([-3,0,-1],w,m)
mblock2 = Parallelepiped([3,0,1],w,m)

objm = RectangularObject()
objm.set_block(mblock1,[-2,0,-1])
objm.set_block(mblock2,[2,0,1])

print(f'B field: {objm.calc_Bfield(p).round(4)} T')
```

```python
print(f'B field: {objm.calc_Bfield([2.2,0,0.5]).round(4)} T')
```

```python
parallelepiped_plot(mblock1)
parallelepiped_plot(mblock2)
plt.xlim(-6,6)
plt.ylim(-6,6)
plt.xlabel('x [m]')
plt.ylabel('z [m]')

x = np.linspace(-5.5,5.5,11)
y = 0
z = np.linspace(-5.5,5.5,11)

plt.grid()

for xp in x:
    for zp in z:
        vec = objm.calc_Bfield([xp,0,zp])
        plt.arrow(xp,zp,vec[0],vec[2],head_width=0.1,head_length=0.1,color='green')

plt.show()
```
